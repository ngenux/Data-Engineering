```
rdd = sc.textFile("dbfs:/FileStore/shared_uploads/dummy.csv")
rdd.collect()
```

```
final_rdd = rdd.zipWithIndex().filter(lambda x: x[1] > 3).map(lambda a :a[0].split(","))
final_rdd.collect()
```

```
columns=final_rdd.collect()[0]
columns
```

```
columns=final_rdd.first() 
data = final_rdd.filter(lambda col: col != columns)  
data.collect()
```
