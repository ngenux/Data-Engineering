```
df = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/customers.csv")
df.show()
```

```
from pyspark.sql.functions import *
df1 = df.withColumn("updatedat",to_date("updatedat",'dd-MM-yyyy')).fillna({"updatedat":'9999-12-31'})
df1.printSchema()
```

```
df1.show()
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import *
df2 = df1.withColumn("rowid",row_number().over(Window.partitionBy("id").orderBy(col("updatedat").desc())))
```

```
df3 = df2.filter("rowid=1")
df3.show()
```
