```
df1 = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/vasachaitanya98@gmail.com/emp__1_.csv")
```

```
df1.show()
```

```
from pyspark.sql.functions import date_format,to_date
df1_csv = df1.withColumn("HIREDATE",to_date("HIREDATE",'dd-MM-yyyy')).fillna({"HIREDATE":'9999-12-31'})
```

```
df1_csv.show()
```

```
df1_csv.\
    withColumn("hire_date", date_format('HIREDATE','yyyyMMdd')).\
    show()
```

```
df1_csv.\
    withColumn("hire_date", date_format('HIREDATE','yyyyMMdd')).\
    coalesce(1).\
    write.\
    partitionBy('hire_date').\
    parquet('dbfs:/FileStore/shared_uploads/partition')
    
```

```
dbutils.fs.ls('dbfs:/FileStore/shared_uploads/partition')
```

```
# partition by month year , day
df1_csv.\
    withColumn('Year', date_format('HIREDATE','yyyy')).\
    withColumn('month', date_format('HIREDATE','MM')).\
    withColumn('day', date_format('HIREDATE','dd')).\
    coalesce(1).\
    write.\
    partitionBy('Year','month','day').\
    parquet('dbfs:/FileStore/shared_uploads/vasachaitanya98@gmail.com/partition')
```
